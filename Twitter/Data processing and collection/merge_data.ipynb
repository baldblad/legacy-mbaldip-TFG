### Short code to merge all tweet files into one dataset file

```python
import pandas as pd
import csv
```

```python
# read all five files and concatenate them all in one single dataframe
df_enAll = pd.concat([pd.read_csv('tweetsSDG_en'+str(x+1)+'.csv') for x in range(5)])
```

```python
df_enAll.info()
display(df_enAll.head())
```

```python
# Check that there are no duplicates (All tweet id should be unique)
print(df_enAll['twid'].nunique()) # unique tweet ids
print(len(df_enAll)) #total number of rows
```

```python
# Remove duplicates using tweet id
df_enAll.drop_duplicates(subset=['twid'], inplace=True)
print(df_enAll['twid'].nunique())
print(len(df_enAll))
```

```python
#Sort by date and reset index
df_enAll = df_enAll.sort_values(by=['created_at'])
df_enAll.reset_index(drop=True, inplace=True)
display(df_enAll.head())
display(df_enAll.tail())
```

```python
# Save the merged dataset
df_enAll.to_csv('tweetsSDG_enALL.csv')
```
